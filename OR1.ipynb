```python
import pandas as pd
```

```python
# 최적화를 위한 데이터 로드
df = pd.read_csv('DreamFinal.csv', encoding='utf-8')
```

```python
df
```

```python
# 값이 3개 이상인 열 제외하고 모든 열 삭제, 즉 개수의 값들이 NaN인 열 제외
df2 = df.dropna(thresh = 3)
```

```python
df2
```

```python
# 데이터 분석을 위해 NaN을 0으로 표시
df3 = df2.fillna(0)
```

```python
df3
```

```python
# 모든 열에 대하여 배달인구, 유동인구, 경쟁업체를 고려하여 최적 입지 도출
for row_index in df3.index:
    if (0.8693 * df3.loc[row_index, '아파트 개수'] + 0.1307 * df3.loc[row_index, '숙박업소 개수'] >= 1
        and 0.5457 * df3.loc[row_index, 'pc방 개수'] + 0.0636 * df3.loc[row_index, '대형마트 개수'] + 0.0285 * df3.loc[row_index, '영화관 개수'] + 0.1520 * df3.loc[row_index, '노래방 개수'] + 0.2096 * df3.loc[row_index, '학교 개수'] >= 1
        and df3.loc[row_index, '경쟁업체 개수'] <= 1):
        print(df3.loc[row_index, 'Unnamed: 0'],df3.loc[row_index, 'polygon'])
```
